```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def opt(alg,file):
    df = pd.read_csv(alg + '/output/freq_'+ file +'.csv')
    coverage = 'Coverage_' + file
    accuracy = 'Accuracy_' + file
    gain = 'Gain' + file
    df[accuracy] = df['Frequency'].cumsum() / df['Frequency'].sum()*100

    df[accuracy] = df[accuracy]
    df[coverage] = df['TripCount'] / df['TripCount'].max() *100
    df[gain] =  df[coverage] - df[accuracy]
    plt.xlabel('Coverage')
    plt.ylabel('Gain')
    
    plt.plot(df[coverage],df[gain])
    plt.axhline(y=0, color='r', linestyle='--')
    
    lowest_point = df.loc[df[gain].idxmin()]
    plt.title(alg + " "+file)
    # Annotating the lowest point
    plt.annotate(f"Lowest: {lowest_point[coverage]}",
             xy=(lowest_point[coverage], lowest_point[gain]),
             xytext=(lowest_point[coverage] + 20, lowest_point[gain] ),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
```

## Binary Search

```python
opt("Binary-search","binary")
```
